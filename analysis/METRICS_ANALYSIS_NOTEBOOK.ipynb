# Анализ метрик HALVITA — демонстрация

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from scipy.stats import pearsonr

def load_sessions(sessions_dir):
    data = []
    for f in Path(sessions_dir).glob("*.json"):
        with open(f) as file:
            s = json.load(file)
            fin = s.get('final_metrics', {})
            data.append({
                'session': s.get('session_id', 'unknown'),
                'IVP': fin.get('IVP', 0),
                'IP': fin.get('IP', 0),
                'INS': fin.get('INS', 0),
                'alpha': fin.get('alpha', 0),
                'beta': fin.get('beta', 0),
                'gamma': fin.get('gamma', 0)
            })
    return pd.DataFrame(data)

df = load_sessions('../sessions/raw')
df.head()
```

```python
plt.hist(df['IVP'], bins=10)
plt.title('Распределение Индекса Свободы')
plt.xlabel('ИВП')
plt.ylabel('Частота')
plt.show()
```

```python
corr = df[['IVP','IP','INS','alpha','beta','gamma']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица метрик')
plt.show()
```

```python
r, p = pearsonr(df['IVP'], df['IP'])
print(f'r = {r:.3f}, p = {p:.4f}')
```
